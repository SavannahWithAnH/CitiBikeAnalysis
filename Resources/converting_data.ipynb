```python
import pandas as pd
```

```python
df = pd.read_csv ('automakers stocks 2010-2022.csv')

df.head(20)     
```

```python
top_ten = ['TSLA', 'TM', 'VOW3.DE', '002594.SZ', 'DAI.DE', 'GM', 'F', 'BMW.DE', 'STLA', 'HMC']

df_filtered = df[df['Symbol'].str.strip().isin(top_ten)]

df_filtered.head()
```

```python
with open('Top 10 Automaker Stocks 2010-2022.json', 'w') as outfile:
    df_filtered.to_json(outfile, orient='records')
```

```python
df_filtered.to_csv ('Top 10 Automaker Stocks 2010-2022.csv')
```
