<h2>Importing Relevant Libraries</h2>

<h3>pprint</h3>
<p>The "pretty-print" library, used to print Python data structures in a well-formatted and more readable way.</p>

<h3>splinter</h3>
<p>A library for automating browser actions, such as visiting URLs, filling in forms, clicking buttons, etc.</p>

<h3>BeautifulSoup (bs4)</h3>
<p>A library used for parsing HTML and XML documents. It creates a parse tree from page source code that can be used to extract data in a hierarchical and more readable manner.</p>

<h3>matplotlib.pyplot (plt)</h3>
<p>A plotting library used for creating static, animated, and interactive visualizations in Python.</p>

<h3>pandas (pd)</h3>
<p>A powerful data manipulation library that offers data structures and operations for manipulating numerical tables and time-series data.</p>

<h3>selenium.webdriver.chrome.options</h3>
<p>The Selenium library is primarily used for automating web applications for testing purposes, but it is certainly not limited to just that.</p>

<h2>Setting Up Selenium WebDriver</h2>
<pre>
<code>
driver_path = '/home/mo/Downloads/chromedriver'
binary_location = '/usr/bin/google-chrome'
options = Options()

</code>
</pre>
<ul>
<li><strong>driver_path:</strong> Defines the path to the Chrome WebDriver executable.</li>
<li><strong>binary_location:</strong> Specifies the location of the Google Chrome binary on your machine.</li>
<li><strong>Options():</strong> Creates a new Options object, which is used to customize and set browser behavior.</li>
<li><strong>options.binary_location:</strong> Sets the binary location of Google Chrome. This is necessary for WebDriver to know where the actual browser is located.</li>
</ul>
